# to fix the problem let's filter out any garbage characters

## first let's see what garbage we can see by looking for anything bigger than 0x7f 

```python
filename = "f:\\CloudStation\\TournamentProject\\RegExport_CoTourn_2018-03-15sample.csv"
with open(filename, 'r+b') as f:
    while True:
        b=f.read(1)
        if not b:
            break
        if( b > b'\x7f'):
            s=b.decode('iso8859_3')
            print("{} {}".format(b,s))

        
```

## next let's re-write the bytes with a '#' or '\x23' 
- notice we're going to change the filename to avoid whacking the orginal

```python
import shutil
shutil.copy("f:\\CloudStation\\TournamentProject\\RegExport_CoTourn_2018-03-15sample.csv","f:\\CloudStation\\TournamentProject\\HasJunkInIt.csv")
filename = "f:\\CloudStation\\TournamentProject\\HasJunkInIt.csv"
print("Cleaning out bad bytes")
with open(filename, 'r+b') as f:
    while True:
        b=f.read(1)
        if not b:
            break
        if( b > b'\x7f'):
            s=b.decode('iso8859_3')
            print("{} {}".format(b,s))
            b=b'\x23'
            f.seek(-1,1)
            f.write(b)
 
print("There should not be any bad bytes left")
with open(filename, 'r+b') as f:
    while True:
        b=f.read(1)
        if not b:
            break
        if( b > b'\x7f'):
            s=b.decode('iso8859_3')
            print("{} {}".format(b,s))
            
print("now read it as a dataframe")            
filebuffer = io.open(filename,'r',encoding='utf-8')
raw_df=pd.read_csv(filebuffer)
print(raw_df)
filebuffer.close()
```
